```python
# 1. 판다스 넘파이 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 2. exam.csv 불러오기
exam = pd.read_csv('exam.csv')
```

```python
# 3. exam 앞에서 7행 불러오기
exam.head(7)
```

```python
# 4. exam 뒤에서 9행 불러오기
exam.tail(9)
```

```python
# 5. exam에서 nclass가 1인 경우만 추출
exam.query('nclass == 1')
```

```python
# 6. exam에서 nclass가 1반이 아닌 경우 추출
exam.query('nclass != 1')
```

```python
# 7. 수학 점수가 50점을 초과한 경우
exam.query('math >= 50')
```

```python
# 8. 영어 점수가 80점 이하인 경우
exam.query('english <= 80')
```

```python
# 9. 1반이면서 수학 점수가 50점 이상인 경우
exam.query('nclass == 1 & math >= 50')
```

```python
# 10. 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query('math >= 90 | english >= 90')
```

```python
# 11. 1,3,4 반에 해당하는 반만 출력
exam.query('nclass == [1,3,4]')
```

```python
# 12. 1반 수학 점수 평균 구하기
exam.query('nclass == 1')['math'].mean()
```
