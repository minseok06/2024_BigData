```python
# 그래프 라이브러리
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 한글폰트 적용
plt.rc('font', family = 'Malgun Gothic')

# 제목
plt.title('월별 판매 실적')
plt.plot(month, sales, color='y', label='단위:개수') # 범례 "단위:개수"
plt.legend(loc = 'lower right')
# x,y 축 제목 월 매출
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
# 그래프 라이브러리
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 한글폰트 적용
plt.rc('font', family = 'Malgun Gothic')

# 제목
plt.title('월별 판매 실적')
plt.bar(month, sales, label='단위:개수') # 범례 "단위:개수"
plt.legend()
# x,y 축 제목 월 매출
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
# 그래프 라이브러리
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 한글폰트 적용
plt.rc('font', family = 'Malgun Gothic')

# 제목
plt.title('월별 판매 실적')
plt.barh(month, sales, label='단위:개수') # 범례 "단위:개수"
plt.legend()
# x,y 축 제목 월 매출
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```
