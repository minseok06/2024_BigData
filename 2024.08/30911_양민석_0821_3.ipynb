```python
import pandas as pd

df_raw = pd.DataFrame({'Var1':[1,2,1],
                    'Var2':[2,3,2]})
df_raw
```

```python
df_raw.shape
```

```python
# 데이터프레임 복사본 만들기.copy()
df_new = df_raw.copy()
df_new
```

# 변수 이름을 변경하는 방법 rename()

```python
# 변수명을 바꾸는 메서드 rename()
df_new = df_new.rename(columns = {'Var2':'V2'})
df_new = df_new.rename(columns = {'Var1':'V1'})
df_new
```

```python
print(df_raw)
print(df_new)
```
