```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# exam.nclass가 1인 경우만 추출
exam.query('nclass==1')
```

```python
# exam.nclass가 2인 경우만 추출
exam.query('nclass==2')
```

```python
# exam.nclass가 3인 경우만 추출
exam.query('nclass==3')
```

```python
# exam.nclass가 4인 경우만 추출
exam.query('nclass==4')
```

```python
# exam.nclass가 5인 경우만 추출
exam.query('nclass==5')
```

```python
# 문제 수학점수가 50점 초과한 경우만 출력
exam.query('math > 50')
```

```python
# 문제 수학점수가 50점 미만인 경우만 출력
exam.query('math < 50')
```

```python
# 1반이면서 수학점수가 50점 이상인경우 추출
exam.query('nclass == 1 & math >= 50')
```

```python
# 2반이면서 영어점수가 80점 이상인 경우
exam.query('nclass == 2 & english >= 80')
```

```python
# 수학점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query('math >= 90 | english >= 90')
```

```python
# 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query('english < 90 | science < 50')
```

```python
# 1,3,5반에 해당하는 값 출력
exam.query('nclass == 1 | nclass == 3 | nclass == 5')
```

```python
# 쉽게 출력하는 방법
exam.query('nclass in [1,3,5]')
```

```python
# nclass가 1인 행 추출해서 nclass1 할당
nclass1 = exam.query('nclass == 1')

# nclass가 2인 행 추출해서 nclass2 할당
nclass2 = exam.query('nclass == 2')
```

```python
# 1반 수학 점수 평균
nclass1['math'].mean()
```

```python
# 2반 수학 점수 평균
nclass2['math'].mean()
```

# 문자 변수를 이용해 조건에 맞는 행 추출하기

```python
df = pd.DataFrame({'sex':['F','M','F','M'],
                  'country':['Korea','China','Japan','USA']})
df
```

```python
# 여자이면서 한국인 데이터 추출 and = 이면서 or = 이거나
# 전체 조건 작은따옴표, 추출할 문자에는 큰따옴표 사용
df.query('sex == "F" & country == "Korea"')
```

```python
# 남자이면서 중국인 데이터 추출
df.query('sex == "M" & country == "China"')
```

```python
var = int(input('반을 입력하세요 : '))
exam.query('nclass == @var')
```
